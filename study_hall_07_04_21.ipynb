```python
# To traverse all of the files in our file_path_list
def create_dataframes(file_path_list,table_names):    
    dataframes=[]

    for x,y in zip(file_path_list,table_names):
        df=pd.read_csv(x, header=None, names=range(5))
        groups = df[0].isin(y).cumsum()
        tables = {v.iloc[0,0]: v.iloc[1:] for k,v in df.groupby(groups)}
        dataframes.append(tables)

    return dataframes
#         print(x,y)
#         print('\n')
```

```python
def create_attempts_df(dataframes): 
    attempts_list = []

    for dictionary in dataframes:
    #     print(dictionary['Attempts'])
        attempts_df = dictionary['Attempts']
        attempts_df.reset_index(drop=True, inplace=True)
        attempts_df.dropna(axis=1, inplace=True)
        attempts_df.columns = attempts_df.iloc[0]
        attempts_df = attempts_df[1:]
        attempts_list.append(attempts_df)
        
    attempts_df = pd.concat(attempts_list).reset_index(drop=True)
    attempts_df['Date'] = pd.to_datetime(attempts_df.Date)
    
    return attempts_df
```
